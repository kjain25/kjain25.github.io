```python
def find_num(y, m):
    if y == 2020:
        return 1
    elif y == 2021:
        return  m + 1
    else:
        return m + 13
```

```python
import pandas as pd
df_main['year'] = pd.to_datetime(df_main['date']).dt.year # create a new column for year
df_main['month'] = pd.to_datetime(df_main['date']).dt.month # create a new column for month
df_main['number'] = df_main.apply(lambda row: find_num(row['year'], row['month']), axis=1)
vaccine_trends = df_main.groupby(['vaccine', 'year', 'month', 'number'])['total_vaccinations'].sum().reset_index().sort_values(['number','vaccine'])
```

```python
vaccine_trends[vaccine_trends.vaccine == 'Pfizer/BioNTech'].to_csv('Pfizer.csv', index = False)
```

```python
vaccine_trends[vaccine_trends.vaccine == 'Moderna'].to_csv('Moderna.csv', index = False)
```

```python
vaccine_trends[vaccine_trends.vaccine == 'Oxford/AstraZeneca'].to_csv('AstraZeneca.csv', index = False)
```

```python
countries = pd.read_csv('countries.csv')
countries
```

```python
df_combined = df_main.groupby(['location', 'vaccine'])['total_vaccinations'].sum().reset_index()
```

```python
df_combined
```

```python
df_combined["vaccine"].unique()
```

```python
df_combined.to_csv('country_numbers.csv', index=False)
```

```python
df_combined= df_combined.pivot(index='location', columns='vaccine', values='total_vaccinations')
df_combined.to_csv('flattened_file.csv') # save the flattened file
df_combined = pd.read_csv('flattened_file.csv')
df_combined
```

```python
final_combined = df_combined.merge(countries, left_on= 'location', right_on='Country')
final_combined
```

```python
df_combined.fillna(0, inplace=True)
```

```python
final_combined.fillna(0, inplace=True)
final_combined
```

```python
final_combined = final_combined.reset_index(drop = True)
final_combined.to_csv('Death_Rates.csv', index = False)
```

```python
death_rates = final_combined[['location', 'Deaths/1M pop']]
death_rates.to_csv('death_rates.csv', index = False)
death_rates
```
